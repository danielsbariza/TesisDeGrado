```python
#@markdown #Importación de las librerías
import pandas as pd
#import numpy as np
from os import mkdir
from google.colab import drive
drive.mount('/content/drive')
```

```python
#@markdown #Importación del dataset
dataset = pd.read_csv("/content/drive/MyDrive/ProyectoDeGrado/Final/Datos.csv")  #lector de archivos csv
datosLen = len(dataset)
```

```python
#@markdown # Selección de datos
#@markdown  Precise el porcentaje de datos a seleccionar: 
porcentajeTrain = 10 #@param {type:"number"}
porcentaje = datosLen*int(porcentajeTrain)/100
Xtrain = []
ytrain = []
paso = round(datosLen/porcentaje)
print(datosLen/porcentaje)
print("Con un paso de:", paso)
posInicial = 0;
#posInicial = int(input('seleccione el punto inicial = '))
print("> procesando datos...")
for k in range(posInicial,datosLen,paso):
  Xtrain.append(dataset.iloc[k , 0:4].values)
  ytrain.append(dataset.iloc[k , 4:7].values)

selec = range(posInicial,datosLen,paso)
datosTest = dataset.drop(selec,axis=0)
Xtest = datosTest.iloc[: , 0:4].values
ytest = datosTest.iloc[: , 4:7].values

X_train = pd.DataFrame(Xtrain)
y_train = pd.DataFrame(ytrain)
X_test = pd.DataFrame(Xtest)
y_test = pd.DataFrame(ytest)
print("> Datos procesados.")
print(len(Xtrain))
```

## **Exportación de datos**


```python
#@markdown # Exportación de datos seleccionados
try:
  mkdir('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Salto/Datos_{}/'.format(porcentajeTrain))
except FileExistsError:
  print("La carpeta ya está creada.")
else: 
  print("carpeta creada con éxito.")
finally:
  X_train.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Salto/Datos_{}/Xtrain.csv'.format(porcentajeTrain))
  y_train.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Salto/Datos_{}/ytrain.csv'.format(porcentajeTrain))
  X_test.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Salto/Datos_{}/Xtest.csv'.format(porcentajeTrain))
  y_test.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Salto/Datos_{}/ytest.csv'.format(porcentajeTrain))
```
