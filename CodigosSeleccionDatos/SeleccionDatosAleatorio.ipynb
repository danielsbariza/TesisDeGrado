```python
#@markdown #Importación de las librerías
import pandas as pd
from os import mkdir
from google.colab import drive
drive.mount('/content/drive')
```

```python
#@markdown #Importación del dataset
dt = pd.read_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos.csv')  #lector de archivos csv
dataset = dt
X_multiple_p = dataset.iloc[:, 0:4].values
y_multiple_p = dataset.iloc[:, 4:7].values
```

```python
#@markdown # Selección de datos
from sklearn.model_selection import train_test_split
#Datos de entrenamiento y prueba
#@markdown  Precise el porcentaje de datos a seleccionar: 
porcentaje = 25 #@param {type:"number"}
print("> procesando datos...")
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_multiple_p, y_multiple_p, test_size=1-(int(porcentaje)/100))
X_train = pd.DataFrame(X_train_p)
y_train = pd.DataFrame(y_train_p)
X_test = pd.DataFrame(X_test_p)
y_test = pd.DataFrame(y_test_p)
print("> Datos procesados.")
```

```python
#@markdown # Exportación de datos seleccionados
mkdir('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Aleatorios/Datos_{}/'.format(porcentaje))
X_train.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Aleatorios/Datos_{}/Xtrain.csv'.format(porcentaje))
y_train.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Aleatorios/Datos_{}/ytrain.csv'.format(porcentaje))
X_test.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Aleatorios/Datos_{}/Xtest.csv'.format(porcentaje))
y_test.to_csv('/content/drive/MyDrive/ProyectoDeGrado/Final/Datos_Aleatorios/Datos_{}/ytest.csv'.format(porcentaje))
```
